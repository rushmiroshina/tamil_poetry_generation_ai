```python
import json
import random
import os

THEMES = ["காதல்", "சமூகம்", "சுதந்திரம்", "இயற்கை", "நெறி"]

INPUT_DIR = "data/clean"
OUTPUT_FILE = "data/fine_tune.jsonl"

def create_instruction(poet, theme):
    return (
        f"{poet} பாணியில் {theme} கருப்பொருளில் "
        f"ஒரு தமிழ் கவிதை வரியை எழுதுக."
    )

with open(OUTPUT_FILE, "w", encoding="utf-8") as out:
    for file in os.listdir(INPUT_DIR):
        poet = file.replace(".txt", "")
        with open(os.path.join(INPUT_DIR, file), encoding="utf-8") as f:
            text = f.read()

        lines = [l.strip() for l in text.split("\n") if len(l.strip()) > 5]

        for line in lines:
            theme = random.choice(THEMES)
            record = {
                "instruction": create_instruction(poet, theme),
                "output": line
            }
            out.write(json.dumps(record, ensure_ascii=False) + "\n")

print("JSONL created:", OUTPUT_FILE)
```
