```python
import os
import json
import random

CLEAN_FOLDER = "clean"
OUTPUT_FILE = "fine_tune.jsonl"

def detect_poet(filename):
    name = filename.lower()
    if "avvai" in name or "ஒளவை" in name:
        return "ஒளவையார்"
    elif "bharathi" in name or "பாரதி" in name or "bharathidasan" in name:
        return "பாரதிதாசன்"
    return None

def build_instruction(poet):
    themes = ["காதல்", "அறம்", "வாழ்க்கை", "இயற்கை"]
    theme = random.choice(themes)

    if poet == "ஒளவையார்":
        return f"{poet} நடைபோல் {theme} கருவில் ஒரு குறள் போன்ற கவிதை வரி எழுதுக."
    else:
        return f"{poet} நடைபோல் {theme} கருவில் ஒரு உணர்ச்சி மிகுந்த கவிதை வரி எழுதுக."

data = []

for file in os.listdir(CLEAN_FOLDER):
    if not file.endswith(".txt"):
        continue

    poet = detect_poet(file)
    if poet is None:
        continue

    with open(os.path.join(CLEAN_FOLDER, file), encoding="utf-8") as f:
        lines = [l.strip() for l in f.readlines() if len(l.strip()) > 15]

    for line in lines:
        if any(x in line for x in ["©", "http", "www", "பதிப்பு"]):
            continue

        sample = {
            "instruction": build_instruction(poet),
            "response": line
        }
        data.append(sample)

print("Total training samples:", len(data))

with open(OUTPUT_FILE, "w", encoding="utf-8") as out:
    for row in data:
        out.write(json.dumps(row, ensure_ascii=False) + "\n")

print("Saved to", OUTPUT_FILE)
```
