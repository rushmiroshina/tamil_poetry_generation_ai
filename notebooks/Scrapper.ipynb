```python
import os
import pandas as pd
import requests

# Folders
os.makedirs("pdf", exist_ok=True)
os.makedirs("unicode", exist_ok=True)

# Read CSV
df = pd.read_csv("projectmadurai_index.csv", encoding="utf-8-sig")
df['PDF'] = df['PDF'].fillna('').apply(lambda x: x.split())
df['Unicode'] = df['Unicode'].fillna('').apply(lambda x: x.split())

# Base URLs
pdf_base = "https://www.projectmadurai.org/pm_etexts/pdf/"
unicode_base = "https://www.projectmadurai.org/pm_etexts/utf8/"

# Track downloaded files
downloaded_pdfs = set()
downloaded_unicode = set()

# Download
for _, row in df.iterrows():
    # PDFs
    for pdf in row['PDF']:
        if pdf and pdf not in downloaded_pdfs:
            url = pdf_base + pdf
            r = requests.get(url)
            if r.status_code == 200:
                with open(os.path.join("pdf", pdf), "wb") as f:
                    f.write(r.content)
                downloaded_pdfs.add(pdf)
    
    # Unicode HTML
    for uni in row['Unicode']:
        if uni and uni not in downloaded_unicode:
            url = unicode_base + uni
            r = requests.get(url)
            if r.status_code == 200:
                with open(os.path.join("unicode", uni), "wb") as f:
                    f.write(r.content)
                downloaded_unicode.add(uni)
```
